```python
import subprocess
import sys

def ask_airbnb(query):
    code = f"""
import asyncio
from airbnb_mcp import search
asyncio.run(search("{query}"))
"""
    result = subprocess.run([sys.executable,'-c',code],capture_output=True, text = True)
    print(result.stdout)
```

```python
query = "show me the best hotels in Mumbai with check in on 31st January 2026 and checkout of 2nd feb 2026"
ask_airbnb(query)
```
